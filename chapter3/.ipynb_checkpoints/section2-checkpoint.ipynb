# 2. 松原形式

**目的**

有限温度における熱平衡Green関数を計算したい $\to$ **松原Green関数**の形式で計算できる

# 2.1 Bloch方程式

熱平衡Green関数の問題点:

   有限温度ではあらゆる励起状態に準粒子が存在してしまう.　$\to$ $t=\pm\infty$の状態が全く異なる状態になるのが必然
   
*[松原武生](https://kotobank.jp/word/%E6%9D%BE%E5%8E%9F%E6%AD%A6%E7%94%9F-1111170)氏のアイデア*: 時間発展演算子と統計演算子が似ている...?
\begin{align}
    \mathcal{U} &= \exp\left(-i\mathcal{H}t\right)\\
    \hat{\rho} &= \exp\left(-\beta\mathcal{H}\right)
\end{align}

## Bloch方程式とSchrodinger方程式

$\beta$の代わりに仮想的な変数$\tau$を使うと, 次のような"Bloch方程式"は成立する.
\begin{align}
    \frac{\partial}{\partial\tau}\hat{\rho}(\tau) = -\mathcal{H}\hat{\rho}(\tau) \tag{3.37}\\
    \hat{\rho}(0)=e^{-0\mathcal{H}}=\mathcal{I}
\end{align}
これは実際に上の統計演算子を代入すればすぐに確認できる.

ここで, 次のような変数変換
\begin{align}
    t \leftrightarrow -i\tau
\end{align}
を$0<\tau<\beta$の範囲で仮想的に導入してみると, 先程のBloch方程式は
\begin{align}
    i\frac{\partial}{\partial t}\hat{\rho}(it) = \mathcal{H}\hat{\rho}(it) \tag{3.38}
\end{align}
$\to$ Schodinger方程式へと変化する.

今は統計演算子のほうを$t\to -i\tau$と変形したが,

場の演算子$\psi$の方の時間軸を$t\to i\tau$として, 
複素平面内で$\pi/2$回転させても話は同じである.

この時間軸回転のことを"**Wick回転**"(Wick's rotation)という.

### 松原演算子

Wick回転を場の演算子に施すことで, Heisenberg演算子は次のような"松原演算子"$\psi^M$の表示へと変換される.
\begin{align}
    \psi(x,t) &= e^{i\mathcal{H}t}\psi(x)e^{-i\mathcal{H}t} &\to \psi^M(x,\tau)&=e^{\mathcal{H}\tau}\psi(x)e^{-\mathcal{H}\tau} \tag{3.40} \\
    \psi^{\dagger}(x,t) &= e^{i\mathcal{H}t}\psi(x)^{\dagger}e^{-i\mathcal{H}t} &\to \overline{\psi}^M(x,\tau)&=e^{\mathcal{H}\tau}\psi(x)^{\dagger}e^{-\mathcal{H}\tau} \tag{3.41}
\end{align}

これらの演算子は, Heisenbergの運動方程式(1.84)を$t\to i\tau$に置き換えた("解析接続"した)運動方程式を満たす
\begin{align}
    \frac{\partial}{\partial\tau}\psi^M(x,\tau) =& \left[\mathcal{H},\psi^M(x,\tau)\right] \tag{3.42} \\
    \frac{\partial}{\partial\tau}\overline{\psi}^M(x,\tau) =& \left[\mathcal{H},\overline{\psi}^M(x,\tau)\right] \tag{3.43}
\end{align}

# 2.2 温度Green関数(松原Green関数)

温度Green関数というものを次のように定義する.
\begin{align}
    G_{\alpha\alpha'}(x,\tau;x',\tau') =& -\left<\mathcal{T}_{\tau}\psi_{\alpha}^M (x,\tau)\overline{\psi}_{\alpha'}^M(x',\tau')\right> \\
    =& -{\rm tr}\left\{e^{-\beta(\mathcal{H}-\Omega)}\mathcal{T}_{\tau}\psi_{\alpha}^M(x,\tau)\overline{\psi}_{\alpha'}^M(x',\tau')\right\} \tag{3.44}
\end{align}
これは（3.10）で定義した熱平衡Green関数を, $t\to i\tau$に置き換えたものであるが,
\begin{align}
    G_{\alpha\beta}(x_1,t_1;x_2,t_2) &= -i\rm{tr}\left(\hat{\rho}\mathcal{T}\psi_{\alpha}(x_1,t_1)\psi_{\beta}^{\dagger}(x_2,t_2)\right) \tag{3.10}\\
\end{align}
時間順序演算子:$\mathcal{T}$については, "**温度順序化演算子**":$\mathcal{T}_{\tau}$として変化しているが, これも$\tau$が大きいものほど前に来るように並び替えるようにする演算子である.

## 松原周波数
温度Green関数は
\begin{align}
    G(x,\tau;x',\tau') =& -{\rm tr}\left\{e^{\beta(\mathcal{H}-\Omega)}\psi^M(x,\tau)\overline{\psi}^M(x',\tau
    )\right\}\\
    =&  -{\rm tr}\left\{e^{\beta(\mathcal{H}-\Omega)}e^{\mathcal{H}\tau}\psi(x)e^{-\mathcal{H}\tau}e^{\mathcal{H}\tau'}\psi^{\dagger}(x'
    )e^{-\mathcal{H}\tau'}\right\}\\
    =&  -{\rm tr}\left\{e^{\beta(\mathcal{H}-\Omega)}e^{\mathcal{H}(\tau-\tau')}\psi(x)e^{-\mathcal{H}(\tau-\tau')}\psi^{\dagger}(x'
    )\right\}\\
\end{align}
と式変形できることから, $\tau-\tau'$に依存することがわかる.

$0<\tau,\tau'<\beta$としていたので, $\tau-\tau
$は$-\beta<\tau-\tau'<\beta$の範囲をとることがわかる.

新しく $\tau \equiv \tau-\tau'$として変数を置くことで,

温度Green関数$G(\tau)$を$2\beta$の周期を持つ関数として扱うことが可能となる.

つまり, 次のようにFourier級数の形で表すことができる.
\begin{align}
    G(\tau) = \frac{1}{\beta}\sum_{n=-\infty}^{\infty} G(\omega_n)e^{-i\omega_n \tau} \tag{3.46}
\end{align}
ここで現れた周波数$\omega_n$を"**松原周波数**"という.

$2\beta$の周期性を持つためには,
\begin{align}
    \omega_n = \frac{\pi n}{\beta}
\end{align}
を選べばよい.

### Fermi/Bose統計

Fermi/Bose統計性を考慮して温度Green関数を考えてみる. 

$\tau<0 \to \tau<\tau'$であるとき, 式(3.44)は
\begin{align}
    G(\tau;\tau') =& -{\rm tr}\left\{e^{-\beta(\mathcal{H}-\Omega)}\mathcal{T}_{\tau}\psi^M(\tau)\overline{\psi}^M(\tau')\right\} \tag{3.44}\\
    =& \pm{\rm tr}\left\{e^{-\beta(\mathcal{H}-\Omega)}\overline{\psi}^M(\tau')\psi^M(\tau)\right\}\\
    =& \pm{\rm tr}\left\{e^{-\beta(\mathcal{H}-\Omega)}e^{\mathcal{H}\tau'}\psi^{\dagger} e^{-\mathcal{H}\tau'}e^{\mathcal{H}\tau}\psi e^{-\mathcal{H}\tau}\right\}\\
    =& \pm{\rm tr}\left\{e^{-\beta(\mathcal{H}-\Omega)}e^{-\mathcal{H}(\tau-\tau')}\psi^{\dagger} e^{\mathcal{H}(\tau-\tau')}\psi\right\}\\
    \to& \pm e^{\beta\Omega}{\rm tr}\left\{e^{-\mathcal{H}(\tau+\beta)}\psi^{\dagger}e^{\mathcal{H}\tau}\psi\right\}
\end{align}
となり, $\tau+\beta$では$\tau>\tau'$となるから
\begin{align}
    G(\tau;\tau') =& -{\rm tr}\left\{e^{-\beta(\mathcal{H}-\Omega)}\mathcal{T}_{\tau}\psi^M(\tau)\overline{\psi}^M(\tau')\right\} \tag{3.44}\\
    =& -{\rm tr}\left\{e^{-\beta(\mathcal{H}-\Omega)}\psi^M(\tau)\overline{\psi}^M(\tau')\right\} \\
    =& -{\rm tr}\left\{e^{-\beta(\mathcal{H}-\Omega)}e^{\mathcal{H}\tau}\psi e^{-\mathcal{H}\tau}e^{\mathcal{H}\tau;}\psi^{\dagger} e^{-\mathcal{H}\tau'}\right\}\\
    =& -{\rm tr}\left\{e^{-\beta(\mathcal{H}-\Omega)}e^{\mathcal{H}(\tau-\tau')}\psi e^{-\mathcal{H}(\tau-\tau')}\psi^{\dagger}\right\}\\
    \to& - e^{\beta\Omega}{\rm tr}\left\{e^{\mathcal{H}(\tau)}\psi e^{-\mathcal{H}(\tau+\beta)}\psi^{\dagger}\right\}\\
    =& \mp G(\tau)
\end{align}
つまり, $\tau$が半周期$\beta$だけ進んだ時に, 松原周波数$\omega_n^{F,B}$は
- Fermi統計: $G(\tau+\beta)=e^{-i\omega_n^F (\tau+\beta)}=e^{-i\omega_n^F\tau}e^{-i\omega_n^F \beta}=-G(\tau)$
- Bose統計 : $G(\tau+\beta)=e^{-i\omega_n^B (\tau+\beta)}=e^{-i\omega_n^B\tau}e^{-i\omega_n^B \beta}=+G(\tau)$

とならなければいけないので,
\begin{align}
    \omega_n^F =& \frac{(2\nu +1)\pi}{\beta} \tag{3.48}\\
    \omega_n^B =& \frac{2\nu \pi}{\beta} \tag{3.49}
\end{align}
と限定される.

最後に$x$についてもFourier変換した形は次の通り
\begin{align}
    G(x,\tau) =& \frac{1}{\beta}\sum_{n=-\infty}^{\infty}\int \frac{d^3 p}{(2\pi)^3} e^{i(px-\omega_{\nu}\tau)}G(p,\omega_{\nu}) \tag{3.50}\\
    G(p,\omega_{\nu}) =& \int_0^{\beta}d\tau \int d^3x e^{-i(px-\omega_{\nu}\tau)}G(x,\tau) \tag{3.51}
\end{align}

## 一般化したKällén-Lehmann表示

温度Green関数でも次のようなKällén-Lehmann表示を得ることができる
\begin{align}
    G(p,\omega_{\nu}) = \left[\frac{1}{2}\right](2\pi)^3\sum_m \sum_n \rho_{n}A_{mn}\delta(p-P_{mn})\left\{ \frac{1\pm e^{-\beta\omega_{mn}}}{i\omega_{\nu}-\omega_{mn}}\right\} \tag{3.52}
\end{align}
熱平衡Green関数は次のように表せた.
\begin{align}
    G^{R}(\mathbf{p},\omega)=\left[\frac{1}{2}\right](2\pi)^3\sum_m \sum_n \rho_n A_{mn}\delta(\mathbf{p}-\mathbf{P}_{mn})\left\{\frac{1\pm e^{-\beta\omega_{mn}}}{\omega-\omega_{mn}+i0}\right\} \tag{3.19}\\
    G^{A}(\mathbf{p},\omega)=\left[\frac{1}{2}\right](2\pi)^3\sum_m \sum_n \rho_n A_{mn}\delta(\mathbf{p}-\mathbf{P}_{mn})\left\{\frac{1\pm e^{-\beta\omega_{mn}}}{\omega-\omega_{mn}-i0}\right\} \tag{3.20}\\
\end{align}

この二つを比較すると
\begin{align}
    G(p,\omega_{\nu}) =& G^{\rm R}(p,i\omega_{\nu});\quad \omega_{\nu}>0 \tag{3.53} \\
    G(p,\omega_{\nu}) =& G^{\rm A}(p,i\omega_{\nu});\quad \omega_{\nu}<0 \tag{3.54}
\end{align}

$\to$ 温度Green関数から熱平衡Green関数へと解析接続することで, retarded,advanced Green関数を手に入れることができる.

# 2.3 温度Green関数に対する摂動級数とダイヤグラムの技法

温度Green関数における**ダイヤグラム**の技法を考えたい.

$\to$ "*松原の相互作用表示*" なるものを導入する

ハミルトニアンが次のように摂動と非摂動の部分に分けられるとする.
\begin{align}
    \mathcal{H} = \mathcal{H}_0 + \mathcal{H}_1
\end{align}

次のように*松原の相互作用表示*
\begin{align}
    \Psi^{\rm M}(x,\tau) = e^{\mathcal{H}_0\tau}\psi(x)e^{-\mathcal{H}_0\tau} \tag{3.55}
\end{align}
を導入すれば
\begin{align}
    \psi^{\rm M}(x,\tau) =& e^{\mathcal{H}\tau}\psi^{\rm M}(x)e^{-\mathcal{H}\tau}\\
    =& e^{\mathcal{H}\tau}e^{-\mathcal{H}_0\tau}\Psi^{\rm M}(x,\tau)e^{\mathcal{H}_0\tau}e^{-\mathcal{H}\tau} \tag{3.56}
\end{align}
と置き換えることができ, また "*松原表示での相互作用表示の虚時間$S$行列*" も考えることができる.

\begin{align}
    \sigma_(\tau_1,\tau_2) = e^{\mathcal{H}_0\tau_1}e^{-\mathcal{H}(\tau_1-\tau_2)}e^{-\mathcal{H}_0\tau_2}
\end{align}
$S$演算子は次の微分方程式を満たす
\begin{align}
    \frac{\partial}{\partial \tau}\sigma(\tau,\tau_2) =& \frac{\partial}{\partial\tau}\left(e^{\mathcal{H}_0\tau}e^{-\mathcal{H}(\tau-\tau_2)}e^{-\mathcal{H}_0\tau_2}\right) \\
    =& \mathcal{H}_0e^{\mathcal{H}_0\tau}e^{-\mathcal{H}(\tau-\tau_2)}e^{-\mathcal{H}_0\tau_2}-e^{\mathcal{H}_0\tau}\mathcal{H}e^{-\mathcal{H}(\tau-\tau_2)}e^{-\mathcal{H}_0\tau_2}\\
    =& -e^{\mathcal{H}_0\tau}\mathcal{H}_1e^{-\mathcal{H}(\tau-\tau_2)}e^{-\mathcal{H}_0\tau_2}\\
    =& \mathcal{H}_1(\tau)\sigma(\tau,\tau_2) \tag{3.60}\\
    \mathcal{H}_1(\tau) =& e^{\mathcal{H}_0\tau}\mathcal{H}_1e^{-\mathcal{H}_0\tau}
\end{align}

この方程式をDyson展開と同様に展開していけば, $\tau_1>\tau_2$として
\begin{align}
    \sigma(\tau_1,\tau_2) =& \int_{\tau_2}^{\tau_1}d\tau \frac{\partial}{\partial\tau}\sigma(\tau,\tau_2)\\
    =& \int_{\tau_2}^{\tau_1}d\tau -\mathcal{H}_1(\tau)\sigma(\tau,\tau_2)\\
    =& \int_{\tau_2}^{\tau_1}d\tau -\mathcal{H}_1(\tau)\int_{\tau_2}^{\tau'}d\tau' -\mathcal{H}_1(\tau')\sigma(\tau',\tau_2)\\
    =& \cdots = \mathcal{T}_{\tau} \exp\left\{-\int_{\tau_2}^{\tau_1} d\tau \mathcal{H}_1(\tau)\right\} \tag{3.61}
\end{align}

### 松原Green関数と松原相互作用表示

続いて, 松原Green関数を松原相互作用表示を用いて表すことを考えてみる.
\begin{align}
    G(x_1,x_2;\tau_1-\tau_2) =& -e^{\beta\Omega}\left[{\rm tr}\left(e^{-\beta\mathcal{H}}\mathcal{T}_{\tau}e^{\mathcal{H}\tau_1}e^{-\mathcal{H}_0\tau_1}\Psi^{\rm M}(\tau_1)e^{\mathcal{H}_0\tau_1}e^{-\mathcal{H}\tau_1}\times e^{\mathcal{H}\tau_2}e^{-\mathcal{H}_0\tau_2}\overline{\Psi}^{\rm M}(\tau_2)e^{\mathcal{H}_0\tau_2}e^{-\mathcal{H}\tau_2}\right)\right]
\end{align}
$\tau_1>\tau_2$のときは上の式は
\begin{align}
    G(x_1,x_2;\tau_1-\tau_2) =& -e^{\beta\Omega}\left[{\rm tr}\left(e^{-\beta\mathcal{H}}e^{\mathcal{H}\tau_1}e^{-\mathcal{H}_0\tau_1}\Psi^{\rm M}(\tau_1)e^{\mathcal{H}_0\tau_1}e^{-\mathcal{H}\tau_1}\times e^{\mathcal{H}\tau_2}e^{-\mathcal{H}_0\tau_2}\overline{\Psi}^{\rm M}(\tau_2)e^{\mathcal{H}_0\tau_2}e^{-\mathcal{H}\tau_2}\right)\right]\\
    =& -e^{\beta\Omega}{\rm tr}\left\{e^{-\beta\mathcal{H}_0}\sigma(\beta,\tau_1)\Psi^{\rm M}(\tau_1)\sigma(\tau_1,\tau_2)\overline{\Psi}^{\rm M}(\tau_2)\sigma(\tau_2,0)\right\}
\end{align}$\tau_1<\tau_2$のときは
\begin{align}
    G(x_1,x_2;\tau_1-\tau_2) =& \pm e^{\beta\Omega}\left[{\rm tr}\left(e^{-\beta\mathcal{H}}e^{\mathcal{H}\tau_2}e^{-\mathcal{H}_0\tau_2}\overline{\Psi}^{\rm M}(\tau_2)e^{\mathcal{H}_0\tau_2}e^{-\mathcal{H}\tau_2}\times e^{\mathcal{H}\tau_1}e^{-\mathcal{H}_0\tau_1}\Psi^{\rm M}(\tau_1)e^{\mathcal{H}_0\tau_1}e^{-\mathcal{H}\tau_1}\right)\right]\\
    =& \pm e^{\beta\Omega}{\rm tr}\left\{e^{-\beta\mathcal{H}_0}\sigma(\beta,\tau_2)\overline{\Psi}^{\rm M}(\tau_2)\sigma(\tau_2,\tau_1)\Psi^{\rm M}(\tau_1)\sigma(\tau_1,0)\right\}
\end{align}
となるので, まとめると
\begin{align}
    G(x_1,x_2;\tau_1-\tau_2) =& -e^{\beta\Omega}{\rm tr}\left\{e^{-\beta\mathcal{H}_0}\mathcal{T}_{\tau}\left(\sigma(\beta,0)\Psi^{\rm M}(\tau_1)\overline{\Psi}^{\rm M}(\tau_2)\right)\right\}\\
    =&-e^{\beta(\Omega-\Omega_0)}{\rm tr}\left\{e^{-\beta(\mathcal{H}_0-\Omega_0)}\mathcal{T}_{\tau}\left(\sigma(\beta,0)\Psi^{\rm M}(\tau_1)\overline{\Psi}^{\rm M}(\tau_2)\right)\right\}\\
    =&-e^{\beta(\Omega-\Omega_0)}\left<\mathcal{T}_{\tau}\left(\sigma(\beta,0)\Psi^{\rm M}(\tau_1)\overline{\Psi}^{\rm M}(\tau_2)\right)\right>_0\\
\end{align}
となり, さらに
\begin{align}
    e^{\beta(\Omega-\Omega_0)} =& \left[e^{\beta(\Omega_0-\Omega)}\right]^{-1} = \left[e^{\beta(\Omega_0-\Omega)}{\rm tr}e^{-\beta(\mathcal{H}-\Omega)}\right]^{-1} = \left[e^{\beta\Omega_0}{\rm tr}e^{-\beta\mathcal{H}}\right]^{-1} = \left[{\rm tr}e^{\beta(\Omega_0-\mathcal{H}_0)}e^{\beta\mathcal{H}_0}e^{-\beta\mathcal{H}}\right]^{-1} = \left[{\rm tr}e^{\beta(\Omega_0-\mathcal{H}_0)}\sigma(\beta,0)\right]^{-1} = \left[\left<\sigma(\beta,0)\right>_0\right]^{-1}
\end{align}
とまとまることから, 温度Green関数は次のような形式で記述できる.
\begin{align}
    G_{\alpha_1,\alpha_2}(x_1,x_2;\tau_1-\tau_2) = -\frac{\left<\mathcal{T}_{\tau}\Psi_{\alpha_1}^{\rm M}(x_1,\tau_1)\overline{\Psi}_{\alpha_2}^{\rm M}(x_2,\tau_2)\sigma(\beta,0)\right>_0}{\left<\sigma(\beta,0)\right>_0} \tag{3.62}
\end{align}
絶対零度の時のcausal Green関数の表式は
\begin{align}
    iG_{\alpha_1,\alpha_2}(x_1,t_1;x_2,t_2)=\frac{\bra{\Psi_0}\mathcal{T}S(\infty,-\infty)\Psi_{\alpha_1}(x_1,t_1)\Psi_{\alpha_2}^{\dagger}(x_2,t_2)\ket{\Psi_0}}{\bra{\Psi_0}S(\infty,-\infty)\ket{\Psi_0}} \tag{2.54}
\end{align}
となり, 二つの式は対応した演算で同等に扱うことが可能となる!

### 松原振動数に関する和

<img src="figure/matsubara_fourier.png" width="600px">
松原周波数は離散的な値をとるので, Fourier変換を計算する際にはその無限級数を解かなくてはならない
- Fermi振動数: $\frac{1}{\beta}\sum_{\nu=-\infty}^{\infty} f(i\omega_{\nu}^{\rm F})$
- Bose振動数 : $\frac{1}{\beta}\sum_{\nu=-\infty}^{\infty} f(i\omega_{\nu}^{\rm B})$

もし, ある関数$f(z)$が$|z|\to\infty$のときに$f(z)\sim|z|^{-(1+\varepsilon)}$であれば, $f(z)$の留数の和のみで表せるようになる.
- Fermi振動数: $\frac{1}{\beta}\sum_{\nu=-\infty}^{\infty} f(i\omega_{\nu}^{\rm F})=-\frac{1}{2}\sum_s\tanh\frac{\beta z_s}{2}{\rm Res}_{z=z_s}f(z)\tag{3.63}$
- Bose振動数 : $\frac{1}{\beta}\sum_{\nu=-\infty}^{\infty} f(i\omega_{\nu}^{\rm B})=-\frac{1}{2}\sum_s\coth\frac{\beta z_s}{2}{\rm Res}_{z=z_s}f(z)\tag{3.64}$

#### 簡単な説明
Fermi統計の場合は次のような松原振動数をとる.
\begin{align}
    \omega_{\nu}^{\rm F} = \frac{(2\nu +1)\pi}{\beta}
\end{align}
ここで, $\tanh$という関数が
\begin{align}
    \tanh\frac{\beta z}{2} = \frac{e^{\beta z}-1}{e^{\beta z}+1}
\end{align}
であるが, これはちょうど $z_{\nu}=i\omega_{\nu}^{\rm F}$を極とする関数となっている.

なので, 図3.4のように円状の経路をとると,
\begin{align}
    \oint dz f(z)\tanh\frac{\beta z}{2} = 2\pi i {\rm Res}\left(f(z)\tanh\frac{\beta z}{2}\right)
\end{align}
という計算となるが, このうち円周部分は$f(z)$が$1/|z|$よりも早く収束することから, 寄与がゼロとなる.

$\tanh$の留数は
\begin{align}
    {\rm Res}\tanh\frac{\beta z}{2} =& \lim_{z\to i\omega_{\nu}^{\rm F}}\left\{(z-i\omega_{\nu}^{\rm F})\tanh\frac{\beta z}{2}\right\} \\
    =& \lim_{z\to i\omega_{\nu}^{\rm F}}\left\{\frac{(z-i\omega_{\nu}^{\rm F})\sinh\frac{\beta z}{2}}{\cosh\frac{\beta z}{2}}\right\} \\
    =& \lim_{z\to i\omega_{\nu}^{\rm F}}\left\{\frac{\sinh\frac{\beta z}{2}+\frac{\beta}{2}(z-i\omega_{\nu}^{\rm F})\cosh\frac{\beta z}{2}}{\frac{\beta}{2}\sinh\frac{\beta z}{2}}\right\} (l'Hôpitalの定理) \\
    =& \lim_{z\to i\omega_{\nu}^{\rm F}}\left\{\frac{2}{\beta} + \coth\frac{\beta z}{2}\right\} \\
    =& \frac{2}{\beta}
\end{align}
となるので, 先程の経路積分は
\begin{align}
    0= 2\pi i {\rm Res}\left(f(z)\tanh\frac{\beta z}{2}\right)
\end{align}
\begin{align}
    &\sum_{\nu=-\infty}^{\infty} f(i\omega_{\nu}^{\rm F}){\rm Res}_{z\to i\omega_{\nu}^{\rm F}}\tanh\frac{\beta z}{2}+\sum_s\tanh\frac{\beta z_s}{2}{\rm Res}_{z=z_s}f(z) = 0 \\
    \Leftrightarrow &\sum_{\nu=-\infty}^{\infty} f(i\omega_{\nu}^{\rm F})\frac{2}{\beta}+\sum_s\tanh\frac{\beta z_s}{2}{\rm Res}_{z=z_s}f(z) = 0 \\
    \Leftrightarrow &\frac{1}{\beta}\sum_{\nu=-\infty}^{\infty} f(i\omega_{\nu}^{\rm F})=-\frac{1}{2}\sum_s\tanh\frac{\beta z_s}{2}{\rm Res}_{z=z_s}f(z) \tag{3.63}
\end{align}
と書けることになる.
