# 3. 線形応答の理論

**目的**

摂動を加えたときに観測量がどのように応答するのかを調べる

# 3.1 線形応答の理論: 久保公式

**線形応答**(linear response) の例
- Hookeの法則: $F=kx$
- Ohmの法則:   $V=IR$

$\uparrow$ これらのような関係を一般化して, 

ある摂動$\mathcal{H}_1(t) = -f(t)\mathcal{B}$を加えたときに, 観測量$\mathcal{A}$がどのように変化する(線形的な関係がある)のかを計算したい.

例:
- $\mathcal{H}_1(t)$: 外部電場, $\mathcal{A}$: 電流
- $\mathcal{H}_1(t)$: 外部磁場, $\mathcal{A}$: 磁化

### 計算
次のようなハミルトニアンを考えてみる.
\begin{align}
    \mathcal{H}(t) = \mathcal{H}_0 + \mathcal{H}_1(t)
\end{align}
ここで
- $\mathcal{H}_0$: 外部からの摂動を除いた全ハミルトニアン(相互作用なども含)
- $\mathcal{H}_1(t)$: 外部から加えた摂動

相互作用表示での統計演算子は
\begin{align}
    \tilde{\rho}(t) =& e^{i\mathcal{H}_0 t}\hat{\rho}(t)e^{-i\mathcal{H}_0t} \tag{3.65} \\
    \hat{\rho}(t) =& \mathcal{U}(t) \hat{\rho}(0) \mathcal{U}^\dagger (t) = e^{-i\mathcal{H}t}\hat{\rho}e^{i\mathcal{H}t} \tag{3.8}
\end{align}
この統計演算子は, 次のようなLiouville方程式を満たす.
\begin{align}
    i\frac{d\tilde{\rho}(t)}{dt} =& i\left(i\mathcal{H}_0e^{i\mathcal{H}_0 t}\hat{\rho}(t)e^{-i\mathcal{H}_0}-e^{i\mathcal{H}_0 t}\hat{\rho}(t)i\mathcal{H}_0e^{-i\mathcal{H}_0t}+ e^{i\mathcal{H}_0 t}\frac{d\hat{\rho}(t)}{dt}e^{-i\mathcal{H}_0t}\right) \\
    =& -\left(\mathcal{H}_0\tilde{\rho}(t)-\tilde{\rho}(t)\mathcal{H}_0+ e^{i\mathcal{H}_0 t}\left(-\mathcal{H}\hat{\rho}(t)+\hat{\rho}(t)\mathcal{H}  \right)e^{-i\mathcal{H}_0t}\right) \\
    =& -\left(e^{i\mathcal{H}_0 t}\left(-\mathcal{H}_1(t)\hat{\rho}(t)+\hat{\rho}(t)\mathcal{H}_1(t)  \right)e^{-i\mathcal{H}_0t}\right) \\
    =& e^{i\mathcal{H}_0 t}\mathcal{H}_1(t)e^{-i\mathcal{H}_0 t}e^{i\mathcal{H}_0 t}\hat{\rho}(t)e^{-i\mathcal{H}_0t}-e^{i\mathcal{H}_0 t}\hat{\rho}(t)e^{-i\mathcal{H}_0 t}e^{i\mathcal{H}_0 t}\mathcal{H}_1(t)e^{-i\mathcal{H}_0t} \\
    =& \left[\tilde{H}_1 (t), \tilde{\rho}(t)\right] \tag{3.66}
\end{align}
積分の形に変換すると,
\begin{align}
    \tilde{\rho}(t) = -i\int_{-\infty}^{t} dt' \left[\tilde{H}_1 (t'), \tilde{\rho}(t') \right] + \tilde{\rho}(-\infty)
\end{align}
ここで, $t=-\infty$においては, 系が平衡状態であったと仮定する.
\begin{align}
    \tilde{\rho}(-\infty) = \hat{\rho}_0 \equiv \exp\left\{\beta(\Omega-\mathcal{H}_0)\right\}
\end{align}
すると摂動に対する密度行列の線形応答は次のように表される.
\begin{align}
    \Delta\tilde{\rho}(t) \equiv \tilde{\rho}(t)-\hat{\rho}_0 =& -i\int_{-\infty}^{t} dt' \left[\tilde{H}_1 (t'), \tilde{\rho}(t') \right] \\
    =& -i\int_{-\infty}^{t} dt' \left[\tilde{H}_1 (t'), \hat{\rho}_0 \right] + \mathcal{O}\left((\mathcal{H}_1)^2\right) \tag{3.68}
\end{align}
※$\tilde{\rho}(t) = \left(\sum (\frac{i}{\hbar}\mathcal{H}_0t)^n\right)\left(\sum (-\frac{i}{\hbar}\mathcal{H}t)^n\right)\hat{\rho}_0\left(\sum (\frac{i}{\hbar}\mathcal{H}t)^n\right)\left(\sum (-\frac{i}{\hbar}\mathcal{H}_0t)^n\right) = \hat{\rho}_0\left(\mathcal{I} + \mathcal{O}\left(\mathcal{H}_1\right)\right)$


実際に観測される観測量というのは平均値$\left<\mathcal{A}\right>_t$として計測できるので, この値を考える.
\begin{align}
    \left<\mathcal{A}\right>_t = {\rm tr}\left\{ \hat{\rho}(t)\mathcal{A} \right\} = {\rm tr}\left\{ \tilde{\rho}(t)\tilde{\mathcal{A}(t)} \right\} \tag{3.67}
\end{align}
$\to$つまり, 統計演算子の線形応答がわかれば, 観測量の線形応答も自動的にわかる.
\begin{align}
    \Delta A(t) =& {\rm tr} \left( \Delta\tilde{\rho}(t) \tilde{\mathcal{A}}(t) \right) \\
    =& -i\int_{-\infty}^{t} dt' {\rm tr}\left(\left[\tilde{H}_1 (t'), \hat{\rho}_0 \right]\mathcal{\tilde{A}}(t) \right) \\
    =& -i\int_{-\infty}^{t} dt' {\rm tr}\left(\tilde{H}_1 (t')\hat{\rho}_0\mathcal{\tilde{A}}(t) - \hat{\rho}_0\tilde{H}_1(t')\mathcal{\tilde{A}}(t) \right) \\
    =& -i\int_{-\infty}^{t} dt' {\rm tr}\left(\hat{\rho}_0\mathcal{\tilde{A}}(t)\tilde{H}_1 (t') - \hat{\rho}_0\tilde{H}_1(t')\mathcal{\tilde{A}}(t) \right) \\
    =& -i\int_{-\infty}^{t} dt'\left<\left[\mathcal{\tilde{A}}(t),\tilde{H}_1 (t')\right]\right> \tag{3.69}
\end{align}

摂動項は次のようにc-数$f(t)$と演算子$\mathcal{B}$として記述すると便利である
\begin{align}
    \mathcal{H}_1(t) = -f(t)\mathcal{B}
\end{align}
例:
- 外部磁場のスピンとの結合: $-\hat{\bf{S}}\cdot\bf{H}(t)$
- ベクトルポテンシャルの電流との結合: $-\frac{1}{c}\hat{\bf{j}}\cdot\bf{A}$

※ただ, いつもこのように$f(t)$と$\mathcal{B}$を見つけ出せるとは限らない.

### 2演算子のretarded Green関数

次のような2つの演算子間の関係性を表すGreen関数を定義する
\begin{align}
    \left<\left(\mathcal{A}(t)\mathcal{B}(t')\right)\right>^R = \frac{1}{i}\left< \left[\mathcal{A}(t),\mathcal{B}(t')\right] \right>\theta(t-t') \tag{3.73}
\end{align}
物理的描像: 時刻$t'$に摂動$\mathcal{B}$が加わる　$\to$　時刻$t$で$\mathcal{A}$が変更を受ける　　それらの相関の様子を表している.

もともと$\mathcal{\tilde{A}}(t) = e^{i\mathcal{H}_0 t}\mathcal{A}e^{-i\mathcal{H}_0t}$であったが, ここでは外部摂動なしのハミルトニアン$\mathcal{H}_0$に対するハイゼンベルグ表示であると考えて$\mathcal{\tilde{A}}=\mathcal{A}$と置いている.

このGreen関数を用いれば,(3.69)は次のように書ける

**久保公式**
\begin{align}
    \Delta A(t) =& -i\int_{-\infty}^{t} dt'(-f(t'))\left<\left[\mathcal{A}(t),\mathcal{B}(t')\right]\right> \\
    =& -\frac{1}{i}\int_{-\infty}^{\infty} dt'f(t')\left<\left[\mathcal{A}(t),\mathcal{B}(t')\right]\right>\theta(t-t') \\
    =& -\int_{-\infty}^{\infty} dt'f(t')\left<\left(\mathcal{A}(t)\mathcal{B}(t')\right)\right>^R \tag{3.74}
\end{align}

系が熱平衡状態であれば, 時刻の差のみに依存性が生じる
\begin{align}
    \left<\left(\mathcal{A}(t)\mathcal{B}(t')\right)\right>^R = \left<\left(\mathcal{A}(0)\mathcal{B}(t'-t)\right)\right>^R = \left<\left(\mathcal{A}(t-t')\mathcal{B}(0)\right)\right>^R \tag{3.75}
\end{align}
それにより, Fourier成分について次の式が成立する.
\begin{align}
    \Delta A(\omega) = -f(\omega)\left<\left<\mathcal{A}\mathcal{B}\right>\right>_{\omega}^R \tag{3.76}
\end{align}
**一般化感受率**を
\begin{align}
    \chi(\omega) = \frac{\Delta A(\omega)}{f(\omega)} \tag{3.77}
\end{align}
と定義すれば,
\begin{align}
    \chi(\omega) = -\left<\left<\mathcal{A}\mathcal{B}\right>\right>_{\omega}^R
\end{align}

# 3.2 揺動散逸定理

### formulation



"**自己相関関数**": 時刻の異なる2点では, 同じ演算子$\mathcal{A}$の間で交換が成立しなくなり, ある程度の変化が生じる.
\begin{align}
    K_A (t,t') = \frac{1}{2}\left< \left\{ \mathcal{A}(t),\mathcal{A}(t') \right\} \right> \tag{3.78}
\end{align}
定常状態であれば, 時刻$t,t'$の差だけで表されるであろうから,
\begin{align}
    K_A(t) = \frac{1}{2}\left< \left\{ \mathcal{A}(t),\mathcal{A}(0) \right\} \right> \tag{3.79}
\end{align}
もし, 時間によって変化が生じなければ(つまり交換関係が成立する場合)
\begin{align}
    K_A(t) = \frac{1}{2}\left< \left\{ \mathcal{A}(t),\mathcal{A}(0) \right\} \right> = \left<\mathcal{A}^2\right>
\end{align}
"**自己分散関数**" のほうが便利な時もある(らしい)
\begin{align}
    K_{\delta A}(t) =& \frac{1}{2}\left< \left\{ \delta\mathcal{A}(t),\delta\mathcal{A}(0) \right\} \right>\\
    =& \frac{1}{2}\left< \left\{ \mathcal{A}(t)-\left<\mathcal{A}\right>,\mathcal{A}(0)-\left<\mathcal{A}\right> \right\} \right> \\
    =& \frac{1}{2}\left< \left\{ \mathcal{A}(t)-\left<\mathcal{A}\right>\right\}\left\{ \mathcal{A}(0)-\left<\mathcal{A}\right>\right\}+\left\{ \mathcal{A}(0)-\left<\mathcal{A}\right>\right\}\left\{ \mathcal{A}(t)-\left<\mathcal{A}\right>\right\} \right> \\
    =& \frac{1}{2}\left< \mathcal{A}(t)\mathcal{A}(0)-\left<\mathcal{A}\right>\left\{\mathcal{A}(t)+\mathcal{A}(0)\right\} + \left<\mathcal{A}\right>^2 + \mathcal{A}(0)\mathcal{A}(t)-\left<\mathcal{A}\right>\left\{\mathcal{A}(t)+\mathcal{A}(0)\right\} + \left<\mathcal{A}\right>^2 \right> \\
    =& K_A(t) + \left<\mathcal{A}\right>^2 - 2 \left<\mathcal{A}\right>^2 = K_A(t) - \left<\mathcal{A}\right>^2 \tag{3.80}
\end{align}

"**ゆらぎのスペクトル密度**": 自己相関関数のFourier変換
\begin{align}
    \left( \mathcal{A}^2 \right)_{\omega} = \int_{-\infty}^{\infty} dt e^{i\omega t}K_A(t) \tag{3.81}
\end{align}
Wiener-Khintchinの定理によると(3.81)の意味は, 

ある振動数の区間$[\omega,\omega+\Delta\omega)$ において, $\mathcal{A}$の揺らぎの区分散逸積分(その区間でどれだけ$\mathcal{A}$が変化しやすいか)である$(\mathcal{A}^2)_\omega\Delta\omega$を計算していることになる.

### Kubo-Martin-Schwinger の恒等式

\begin{align}
    \left<\mathcal{A}(t)\mathcal{B}(0)\right> =& {\rm tr}\left(e^{\beta(\Omega-\mathcal{H})}e^{i\mathcal{H}t}\mathcal{A}(0)e^{-i\mathcal{H}t}\mathcal{B}(0)\right)\\
    =& {\rm tr}\left(e^{\beta\Omega}\mathcal{B}(0)e^{-\beta\mathcal{H}}e^{i\mathcal{H}t}\mathcal{A}(0)e^{-i\mathcal{H}t}\right) \\
    =& {\rm tr}\left(e^{-\beta\mathcal{H}}e^{\beta\Omega}\mathcal{B}(0)e^{-\beta\mathcal{H}}e^{i\mathcal{H}t}\mathcal{A}(0)e^{-i\mathcal{H}t}e^{\beta\mathcal{H}}\right) \\
    =& {\rm tr}\left(e^{\beta(\Omega-\mathcal{H})}\mathcal{B}(0)e^{i\mathcal{H}(t+i\beta)}\mathcal{A}(0)e^{-i\mathcal{H}(t+i\beta)}\right) \\
    =& \left<\mathcal{B}(0)\mathcal{A}(t+i\beta)\right> \tag{3.82}
\end{align}
この関係をつかってスペクトル密度を次のように書き直すことができる.
\begin{align}
    \left( \mathcal{A}^2 \right)_{\omega} =& \int_{-\infty}^{\infty} dt e^{i\omega t}K_A(t)
    = \int_{-\infty}^{\infty} dt e^{i\omega t}\frac{1}{2}\left< \left\{ \mathcal{A}(t),\mathcal{A}(0) \right\} \right> \\
    =& \int_{-\infty}^{\infty} dt e^{i\omega t}\frac{1}{2}\left(\left<\mathcal{A}(t),\mathcal{A}(0)\right> + \left<\mathcal{A}(0),\mathcal{A}(t)\right> \right)\\
    =& \frac{1}{2}\int_{-\infty}^{\infty} dt e^{i\omega t}\left<\mathcal{A}(t),\mathcal{A}(0)\right> + \frac{1}{2}\int_{-\infty}^{\infty} dt\left<\mathcal{A}(0),\mathcal{A}(t)\right>\\
    =& \frac{1}{2}\int_{-\infty}^{\infty} dt e^{i\omega t}\left<\mathcal{A}(0),\mathcal{A}(t+i\beta)\right> + \frac{1}{2}\int_{-\infty}^{\infty} dt\left<\mathcal{A}(0),\mathcal{A}(t)\right>\\
    =& \frac{1}{2}\int_{-\infty}^{\infty} dt e^{i\omega (t-i\beta)}\left<\mathcal{A}(0),\mathcal{A}(t)\right> + \frac{1}{2}\int_{-\infty}^{\infty} dt\left<\mathcal{A}(0),\mathcal{A}(t)\right>\\
    =& \frac{1}{2}e^{\beta\omega}\int_{-\infty}^{\infty} dt e^{i\omega t}\left<\mathcal{A}(0),\mathcal{A}(t)\right> + \frac{1}{2}\int_{-\infty}^{\infty} dt\left<\mathcal{A}(0),\mathcal{A}(t)\right>\\
    =& \frac{1}{2}(1+e^{\beta\omega})\int_{-\infty}^{\infty} dt e^{i\omega t}\left<\mathcal{A}(0),\mathcal{A}(t)\right> \tag{3.83}
\end{align}
反交換と同様に交換関係における場合も計算できる.
\begin{align}
    \left(\left[A,A \right]\right)_{\omega} \equiv& \int_{-\infty}^{\infty} dt e^{i\omega t}\frac{1}{2}\left< \left[ \mathcal{A}(t),\mathcal{A}(0) \right] \right> \\
    =& \int_{-\infty}^{\infty} dt e^{i\omega t}\frac{1}{2}\left(\left<\mathcal{A}(t),\mathcal{A}(0)\right> - \left<\mathcal{A}(0),\mathcal{A}(t)\right> \right)\\
    =& \frac{1}{2}(e^{\beta\omega}-1)\int_{-\infty}^{\infty} dt e^{i\omega t}\left<\mathcal{A}(0),\mathcal{A}(t)\right> \tag{3.84}
\end{align}

すると, 揺らぎのスペクトル密度と交換子のスペクトル密度に以下の関係性があることが示せる.
\begin{align}
    &\frac{1}{2}\int_{-\infty}^{\infty} dt e^{i\omega t}\left<\mathcal{A}(0),\mathcal{A}(t)\right>  = \frac{\left( \mathcal{A}^2 \right)_{\omega}}{(e^{\beta\omega}+1)} = \frac{\left(\left[A,A \right]\right)_{\omega}}{(e^{\beta\omega}-1)}\\
    &\Leftrightarrow \frac{\left( \mathcal{A}^2 \right)_{\omega}}{(e^{\beta\omega}+1)} = \frac{\left(\left[A,A \right]\right)_{\omega}}{(e^{\beta\omega}-1)} \\
    &\Leftrightarrow \left( \mathcal{A}^2 \right)_{\omega}= \frac{e^{\beta\omega}+1}{e^{\beta\omega}-1}\left(\left[A,A \right]\right)_{\omega} = \frac{1}{2}\coth\frac{\beta\omega}{2}\left(\left[A,A \right]\right)_{\omega} \tag{3.85}
\end{align}



### 揺動散逸定理
一方で, 次のような関係も書ける.
\begin{align}
    \left(\left[A,A \right]\right)_{\omega} =& \int_{0}^{\infty} dt e^{i\omega t}\frac{1}{2}\left< \left[ \mathcal{A}(t),\mathcal{A}(0) \right] \right> + \int_{-\infty}^{0} dt e^{i\omega t}\frac{1}{2}\left< \left[ \mathcal{A}(t),\mathcal{A}(0) \right] \right> \\
    =& \int_{0}^{\infty} dt e^{i\omega t}\frac{1}{2}\left< \left[ \mathcal{A}(t),\mathcal{A}(0) \right] \right> + \int_{\infty}^{0}(-dt) e^{-i\omega t}\frac{1}{2}\left< \left[ \mathcal{A}(-t),\mathcal{A}(0) \right] \right> \\
    =& \int_{0}^{\infty} dt e^{i\omega t}\frac{1}{2}\left< \left[ \mathcal{A}(t),\mathcal{A}(0) \right] \right> + \int_{0}^{\infty}dt e^{-i\omega t}\frac{1}{2}\left< \left[ \mathcal{A}(0),\mathcal{A}(t) \right] \right> (定常状態)\\
    =& \int_{0}^{\infty} dt e^{i\omega t}\frac{1}{2}\left< \left[ \mathcal{A}(t),\mathcal{A}(0) \right] \right> - \int_{0}^{\infty}dt e^{-i\omega t}\frac{1}{2}\left< \left[ \mathcal{A}(t),\mathcal{A}(0) \right] \right> \\
    =& -2{\rm Im}\int_{0}^{\infty} dt e^{-i\omega t}\frac{1}{2}\left< \left[ \mathcal{A}(t),\mathcal{A}(0) \right] \right> \\
    =& -2{\rm Im}\left< \left< \mathcal{A}\mathcal{A} \right> \right>_{\omega}^R \tag{3.86}
\end{align}

以上までの式を整理すると,
\begin{align}
    \left( \mathcal{A}^2 \right)_{\omega} = \frac{1}{2}\coth\frac{\beta\omega}{2}\left(\left[A,A \right]\right)_{\omega} = -\coth\frac{\beta\omega}{2}{\rm Im}\left< \left< \mathcal{A}\mathcal{A} \right> \right>_{\omega}^R = \coth\frac{\beta\omega}{2}{\rm Im}\chi(\omega) \tag{3.87}
\end{align}
これは"**揺動散逸定理**"(Callen-Welton's formula) と呼ばれる関係である.

$\to$ 観測量の揺らぎが一般化感受率の虚部に比例する

## 線形回路における電流と電圧の揺らぎ: Nyquistの定理

<img src="figure/johnson-nyquist_noize.png" width="500px">

上の図のような回路を例に, 揺動散逸定理を適用してみる.

電流の演算子: $\mathcal{J}$　とおく

十分に低い周波数(周波数$0$であれば, 定常状態と等しい)であれば, 回路のどこでも同じ電流値を持つ.
\begin{align}
    \left<\mathcal{J}(t)\right> = J(t)
\end{align}
回路に外部からの起電力$W(t)$が働く場合(これが外部からの摂動), 単位時間当たりのエネルギー散逸は次式のとおり
\begin{align}
    Q = JW = \left<\mathcal{J}\right>W
\end{align}

一般化力は次のように表せる
\begin{align}
    Q = \mathcal{\dot{E}} =& \left<\mathcal{\dot{H}}\right> = q(t)\left<\mathcal{B}\right> \tag{3.71}\\
    &\Rightarrow q(t) = -\dot{f}(t) \tag{3.72}
\end{align}
つまり今回の場合は,
\begin{align}
    &\dot{f} = -W \\
    \Leftrightarrow& W(\omega)=\int_{-\infty}^{\infty}dt e^{i\omega t}W(t) = \left[-e^{i\omega t}f(t)\right]_{-\infty}^{\infty}+\int_{-\infty}^{\infty}dt (i\omega)e^{i\omega t}f(t) = i\omega f(\omega)
\end{align}
また一般化感受率は$\chi(\omega) = \frac{\Delta A(\omega)}{f(\omega)} (3.77)$で表されることから,
\begin{align}
    J(\omega) = \chi(\omega)f(\omega)
\end{align}
と書くことができ, また電気回路で用いられるインピーダンス$Z$を用いて
\begin{align}
    J(\omega) = \frac{W(\omega)}{Z(\omega)} = \frac{i\omega f(\omega)}{Z(\omega)}
\end{align}
とも書けるので,
\begin{align}
    \chi(\omega)=\frac{i\omega}{Z(\omega)}=\frac{i\omega}{R(\omega)+iY(\omega)} \tag{3.88}
\end{align}
\begin{align}
    {\rm Im}\chi(\omega) = {\rm Im}\frac{i\omega(R(\omega)+iY(\omega))}{\left|Z(\omega)\right|^2} = \frac{\omega}{\left|Z(\omega)\right|^2}R(\omega) \tag{3.89}
\end{align}
最後に揺動散逸定理に適用すると, 電流の揺らぎが計算できる.
\begin{align}
    \left(J^2\right)_{\omega} =& \frac{\omega}{\left|Z(\omega)\right|^2}R(\omega)\coth\frac{\beta\omega}{2} \\
       =& {\rm Re}\left\{Z(\omega)^{-1}\right\}\omega\coth\frac{\beta\omega}{2} \tag{3.90}
\end{align}
同様に電圧揺らぎを計算する. 電圧の演算子を$\mathcal{W}$と置くと,
\begin{align}
    Q  = \left<\mathcal{W}\right>J, \quad \dot{f} = -J, \quad  J(\omega) = i\omega f(\omega)
\end{align}
一般化感受率は
\begin{align}
    W(\omega) = \chi(\omega)f(\omega)
\end{align}
インピーダンス$Z$を用いて
\begin{align}
    W(\omega) = J(\omega)Z(\omega) = i\omega f(\omega)Z(\omega)
\end{align}
\begin{align}
    {\rm Im}\chi(\omega) = {\rm Im}\left\{i\omega(R(\omega)+iY(\omega))\right\} = \omega R(\omega)
\end{align}
よって, 誘導散逸定理より
\begin{align}
    \left(W^2\right)_{\omega} =& \omega R(\omega)\coth\frac{\beta\omega}{2} \\
       =& {\rm Re}\left\{Z(\omega)\right\}\omega\coth\frac{\beta\omega}{2} \tag{3.91}
\end{align}

最後に古典極限$T>>\omega$において考えてみると, $\beta\omega<<1$となるので
\begin{align}
    \coth\frac{\beta\omega}{2} = \frac{\exp(\beta\omega/2) + \exp(-\beta\omega/2)}{\exp(\beta\omega/2) - \exp(-\beta\omega/2)} \approx \frac{2}{\beta\omega} = 2T\omega^{-1}
\end{align}
従って, $G=1/R$とすれば,
\begin{align}
    \left(J^2\right)_{\omega} =& {\rm Re}\left\{Z(\omega)^{-1}\right\}\omega\coth\frac{\beta\omega}{2} \\
    =& 2TG(\omega) \tag{3.92}\\
    \left(W^2\right)_{\omega} =& {\rm Re}\left\{Z(\omega)\right\}\omega\coth\frac{\beta\omega}{2} \\
    =& 2TR(\omega) \tag{3.93}
\end{align}
以上の定理を"**Nyquistの定理**"という.

$\to$熱によって, 電流や電圧がゆらぐ(熱雑音)ことがわかる.
